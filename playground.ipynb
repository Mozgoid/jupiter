```python
import math
import random

print("hello")

# давай напишем генерацию данных что бы по ним построить график
# я задаю параметры генерации данных, например выбираю из разных функций, их параметры, их количество

def generate_data(func, params, n):
    return [func(x, *params) for x in range(n)]

def linear(x, a, b):
    return a*x + b

def quadratic(x, a, b, c):
    return a*x**2 + b*x + c

def cubic(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d

def exponential(x, a, b):
    return a * b**x

def logarithmic(x, a, b):
    return a * math.log(x+1, b)

def sinusoidal(x, a, b, c, d):
    return a*math.sin(b*x + c) + d

def random(x):
    return random.randint(0, 100)
```

```python
# теперь наверное нужна функция для визуализации данных

import matplotlib.pyplot as plt

def plot_data(data):
    plt.plot(data)
    plt.show()
```

```python
# давай потестим графики

data = generate_data(linear, (2, 3), 100)
plot_data(data)
```

```python

data = generate_data(quadratic, (1, 2, 3), 100)
plot_data(data)
```

```python

data = generate_data(exponential, (1, 2), 100)
plot_data(data)
```

```python

data = generate_data(logarithmic, (1, 2), 100)
plot_data(data)
```

```python

data = generate_data(sinusoidal, (1, 2, 3, 4), 100)
plot_data(data)
```
